2. What is an interpreter in Python?
 - A Python interpreter is a software application that reads and executes Python code. It acts as a translator, converting your human-readable Python script into low-level instructions that your computer's hardware can understand.

3. What are predefined Keywords in Python?
 - Predefined keywords in Python are reserved words that have a fixed, specific meaning and cannot be used as identifiers (such as variable names, function names, or class names). These words establish the syntax, control flow, and structural rules of the Python language.

4. Can keywords be used as variable names?
 -  No, you cannot use strict Python keywords as variable names. Keywords are reserved words that define the language's grammar and syntax. If you try to assign a value to a strict keyword (such as if, for, class, or def), the Python parser will reject the code and raise a SyntaxError before the program even runs.

5.  What is mutability in Python?
 -  In Python, mutability refers to the capability of an object to change or modify its internal state or content after it has been created. When you modify a mutable object, Python updates the value directly at its existing memory address (in-place modification) without creating a new object. Conversely, immutable objects cannot be altered once defined; any attempt to modify them forces Python to create a brand-new object in memory.

6. Why are lists mutable, but tuples are immutable?
 -  In Python, lists are mutable and tuples are immutable because they are designed for fundamentally different purposes, semantic use cases, and performance optimizations.

7. What is the difference between “==” and “is” operators in Python?
 -  In Python, the == operator checks for value equality, whereas the is operator checks for object identity.This means == evaluates whether two objects contain the same data, while is evaluates whether two variables point to the exact same memory address.

8. What are logical operators in Python?
 -  Logical operators in Python are reserved keywords used to combine or modify conditional statements. They evaluate expressions and typically return a Boolean value (True or False). Python features exactly three logical operators: and, or, and not.

9. What is type casting in Python?
 -  Type casting in Python is the process of converting a variable or value from one data type to another. It ensures data compatibility when performing operations, such as converting a numeric string from user input into an actual number for mathematical calculations. Python handles this through two different methods: implicit and explicit conversion.

10. What is the difference between implicit and explicit type casting?
 -  The primary difference is that implicit type casting is performed automatically by the Python interpreter, while explicit type casting requires manual intervention by the programmer using built-in functions.

11. What is the purpose of conditional statements in Python?
 -  The purpose of conditional statements in Python is to control the flow of execution by allowing a program to make decisions and execute specific blocks of code based on whether a condition is True or False.

12. How does the elif statement work?
 -  In Python, the elif statement (short for "else if") allows you to check multiple sequential conditions. It only evaluates if all the preceding if and elif conditions in the same logical block have turned out to be False. How It Works Top-to-Bottom Evaluation: Python reads an if-elif-else block sequentially from top to bottom.First Match Wins: As soon as Python finds a condition that is True, it runs that specific block of code. Exits the Block: After running the matching code, Python completely skips all remaining elif or else statements in that block—even if subsequent conditions are also technically true.

13. What is the difference between for and while loops?
  -  The primary difference is that a for loop iterates over a predefined sequence or iterable object, while a while loop repeatedly executes code as long as a specific condition evaluates to True.

14. Describe a scenario where a while loop is more suitable than a for loop?
 -  A while loop is more suitable than a for loop when the exact number of iterations is unknown in advance and the loop must continue running until a specific dynamic condition changes. While a for loop is built to iterate over a predetermined sequence or range, a while loop excels when the exit criteria depend on external events, user inputs, or real-time state changes. A classic example where a while loop is necessary is forcing a user to provide valid data. You cannot use a for loop because you have no way of predicting how many times the user will type an invalid answer before finally entering the correct one.

```python
##6. Write a function to demonstrate the behavior of mutable and immutable arguments?
# Immutable argument example (integer)
def modify_immutable(num):
    print("Inside function before change:", num)
    num += 10
    print("Inside function after change:", num)

# Mutable argument example (list)
def modify_mutable(lst):
    print("Inside function before change:", lst)
    lst.append(100)
    print("Inside function after change:", lst)

# Main program
x = 20
print("Before function call (immutable):", x)
modify_immutable(x)
print("After function call (immutable):", x)

print()

numbers = [1, 2, 3]
print("Before function call (mutable):", numbers)
modify_mutable(numbers)
print("After function call (mutable):", numbers)
Output
Before function call (immutable): 20
Inside function before change: 20
Inside function after change: 30
After function call (immutable): 20

Before function call (mutable): [1, 2, 3]
Inside function before change: [1, 2, 3]
Inside function after change: [1, 2, 3, 100]
After function call (mutable): [1, 2, 3, 100]
```

```python
##8. Write a program to demonstrate the use of logical operators.

# Input age and citizenship status
age = int(input("Enter your age: "))
citizen = input("Are you a citizen? (yes/no): ").lower()

# Using AND operator
if age >= 18 and citizen == "yes":
    print("Eligible to vote.")
else:
    print("Not eligible to vote.")

# Using OR operator
if age < 18 or citizen != "yes":
    print("At least one voting condition is not satisfied.")

# Using NOT operator
if not (citizen == "yes"):
    print("You are not a citizen.")
Sample Output
Enter your age: 20
Are you a citizen? (yes/no): yes

Eligible to vote.
Logical Operators Used
and → Returns True if both conditions are true.
or → Returns True if at least one condition is true.
not → Reverses the logical result (True becomes False and vice versa).
```

```python
##10. Write code to demonstrate type casting with list elements.

# List of numbers stored as strings
string_list = ["10", "20", "30", "40", "50"]

print("Original List:", string_list)
print("Type of first element:", type(string_list[0]))

# Convert string elements to integers
int_list = [int(item) for item in string_list]

print("\nList after type casting to integers:", int_list)
print("Type of first element:", type(int_list[0]))

# Convert integer elements to floats
float_list = [float(item) for item in int_list]

print("\nList after type casting to floats:", float_list)
print("Type of first element:", type(float_list[0]))
Sample Output
Original List: ['10', '20', '30', '40', '50']
Type of first element: <class 'str'>

List after type casting to integers: [10, 20, 30, 40, 50]
Type of first element: <class 'int'>

List after type casting to floats: [10.0, 20.0, 30.0, 40.0, 50.0]
Type of first element: <class 'float'>
Explanation
int(item) converts each string element to an integer.
float(item) converts each integer element to a float.
List comprehension ([int(item) for item in string_list]) applies the conversion to every element in the list.
```

```python
##13. Write a Python program to find the sum of all even numbers between 1 and 50.

# Find the sum of all even numbers between 1 and 50

total = 0

for num in range(1, 51):
    if num % 2 == 0:
        total += num

print("Sum of all even numbers between 1 and 50 is:", total)
Output
Sum of all even numbers between 1 and 50 is: 650
Alternative (More Efficient)
total = sum(range(2, 51, 2))
print("Sum of all even numbers between 1 and 50 is:", total)
Explanation
num % 2 == 0 checks whether a number is even.
total += num adds the even number to the running sum.
The final sum of all even numbers from 1 to 50 is 650.
```

```python
##14. Write a program to reverse a string using a while loop.

# Program to reverse a string using a while loop

text = input("Enter a string: ")

reverse = ""
i = len(text) - 1

while i >= 0:
    reverse += text[i]
    i -= 1

print("Reversed string:", reverse)
Sample Output
Enter a string: Python
Reversed string: nohtyP
Explanation
len(text) - 1 gives the index of the last character.
The while loop runs from the last character to the first.
Each character is added to reverse.
i -= 1 moves to the previous character until the entire string is reversed.
```

```python
##15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

# Program to calculate the factorial of a number using a while loop

num = int(input("Enter a number: "))

factorial = 1
i = 1

while i <= num:
    factorial *= i
    i += 1

print("Factorial of", num, "is", factorial)
Sample Output
Enter a number: 5
Factorial of 5 is 120
Explanation
The factorial of a number n is the product of all positive integers from 1 to n.
The while loop multiplies factorial by each number from 1 to num.
For example:
5! = 5 × 4 × 3 × 2 × 1 = 120
Handling Negative Numbers
num = int(input("Enter a number: "))

if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    factorial = 1
    i = 1

    while i <= num:
        factorial *= i
        i += 1

    print("Factorial of", num, "is", factorial)

This version checks for negative input before calculating the factorial.
```
